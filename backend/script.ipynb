```python
import random
import json
from datetime import datetime, timedelta
```

```python
def generate_random_date():
    start_date = datetime(2015, 1, 1)
    end_date = datetime(2024, 12, 31)
    random_days = random.randint(0, (end_date - start_date).days)
    random_date = start_date + timedelta(days=random_days)
    return random_date.strftime("%Y/%m/%d")
```

```python
# create random posts
posts = []
while len(posts) < 10_000:
    srcI = 0
    while True:
        if random.random() < 0.3:
            break
        srcI += 1
        if srcI >= len(sources):
            srcI = 0
    src = sources[srcI]

    creator = random.choice(creators)
    for j in range(random.randint(5, 150)):
        postTags = []
        for k in range(random.randint(1, 10)):
            postTags.append(random.choice(tags))
        postTags = list(set(postTags))
        post = {
            'source': src,
            'source_id': len(posts),
            'creator': creator,
            'tags': postTags,
            'date_uploaded': generate_random_date(),
            'date_added': generate_random_date(),
            'uri': '...',
        }
        post['id'] = post['source'] + '-' + str(post['source_id'])
        posts.append(post)
random.shuffle(posts)
```

```python
with open('posts.json', 'w') as f:
    json.dump(posts, f, indent=4)
```
